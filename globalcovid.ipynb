```python
df = df.rename(columns={
    'Date': 'date',
    'Country/Region': 'location',
    'Confirmed': 'total_cases',
    'Deaths': 'total_deaths',
})
```

```python
df = df.sort_values(['location', 'date'])
df['new_cases'] = df.groupby('location')['total_cases'].diff().fillna(0)
```

```python
rename_map = {}
col_list = [c.lower() for c in df.columns]
if 'total_cases' not in col_list:
    for alt in ['confirmed', 'cases', 'totalconfirmed', 'total_cases']:
        if alt in col_list:
            rename_map[ df.columns[col_list.index(alt)] ] = 'total_cases'
            break
if 'total_deaths' not in col_list:
    for alt in ['deaths', 'totaldeaths', 'death']:
        if alt in col_list:
            rename_map[ df.columns[col_list.index(alt)] ] = 'total_deaths'
            break
if 'location' not in col_list:
    for alt in ['country', 'country_region', 'country/region']:
        if alt in col_list:
            rename_map[ df.columns[col_list.index(alt)] ] = 'location'
            break
if 'date' not in col_list:
    for alt in ['day', 'date_reported', 'last_update']:
        if alt in col_list:
            rename_map[ df.columns[col_list.index(alt)] ] = 'date'
            break
df.rename(columns=rename_map, inplace=True)
df['date'] = pd.to_datetime(df['date'])
if 'new_cases' not in df.columns:
    df = df.sort_values(['location', 'date'])
    df['new_cases'] = df.groupby('location')['total_cases'].diff().fillna(0)
```

```python
print(df.columns.tolist())
```

```python
selected_countries = ['India', 'United States', 'Brazil', 'Russia', 'United Kingdom']
filtered_df = df[df['location'].isin(selected_countries)]
plt.figure(figsize=(12,6))
for country in selected_countries:
    country_data = filtered_df[filtered_df['location'] == country]
    plt.plot(country_data['date'], country_data['new_cases'], label=country)
plt.title("New COVID-19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
top_deaths = df.groupby('location')['total_deaths'].max().sort_values(ascending=False).head(10)
top_recoveries = df.groupby('location')['total_recovered'].max().sort_values(ascending=False).head(10)
comparison = pd.DataFrame({
    'Deaths': top_deaths,
    'Recoveries': top_recoveries
}).dropna()
comparison.plot(kind='bar', figsize=(12,6), color=['red', 'green'])
plt.title("Total Deaths vs Recoveries (Top Countries)")
plt.xlabel("Country")
plt.ylabel("People")
plt.xticks(rotation=45)
plt.show()
```

```python
top_recoveries = df.groupby('location')['total_recovered'].max().sort_values(ascending=False).head(10)
```

```python
df['estimated_recoveries'] = df['total_cases'] - df['total_deaths']
top_deaths = df.groupby('location')['total_deaths'].max().sort_values(ascending=False).head(10)
top_recoveries = df.groupby('location')['estimated_recoveries'].max().sort_values(ascending=False).head(10)
comparison = pd.DataFrame({
    'Deaths': top_deaths,
    'Estimated Recoveries': top_recoveries
}).dropna()
comparison.plot(kind='bar', figsize=(12,6), color=['red', 'green'])
plt.title("Total Deaths vs Estimated Recoveries (Top Countries)")
plt.xlabel("Country")
plt.ylabel("People")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
global_trend = df.groupby('date')[['total_cases', 'total_deaths']].sum()
plt.figure(figsize=(12,6))
plt.plot(global_trend.index, global_trend['total_cases'], label='Total Cases', color='blue')
plt.plot(global_trend.index, global_trend['total_deaths'], label='Total Deaths', color='red')
plt.title("Global COVID-19 Growth Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
india_df = df[df['location'] == 'India']
plt.figure(figsize=(12,6))
plt.plot(india_df['date'], india_df['total_cases'], label='India Total Cases', color='orange')
plt.title("Indias COVID-19 Case Growth Over Time")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
latest = df.sort_values("date").drop_duplicates("location", keep="last")
latest['death_rate (%)'] = (latest['total_deaths'] / latest['total_cases']) * 100
significant = latest[latest['total_cases'] > 100000]
top_death_rates = significant.sort_values("death_rate (%)", ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_death_rates['death_rate (%)'], y=top_death_rates['location'], palette='Reds')
plt.title("Top 10 Countries by COVID-19 Death Rate")
plt.xlabel("Death Rate (%)")
plt.ylabel("Country")
plt.show()
```
